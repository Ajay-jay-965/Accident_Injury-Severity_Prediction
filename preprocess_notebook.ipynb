```python
def one_hot_top_x(df, column, top_x_labels):
    for label in top_x_labels:
        df[column+'_'+label]=np.where(df[column]==label, 1, 0)
df=pd.read_csv("table1.csv", usecols = ['AGE','SEX','PERSON_TYPE','SEATING_POSITION','RESTRAINT_SYSTEM-USE','AIR_BAG_AVAILABILITY/DEPLOYMENT','EJECTION','NON_MOTORIST_LOCATION','POLICE_REPORTED_ALCOHOL_INVOLVEMENT','ALCOHOL_TEST_TYPE','ALCOHOL_TEST_RESULT','POLICE-REPORTED_DRUG_INVOLVEMENT','DRUG_TEST_TYPE','DRUG_TEST_RESULTS_(1_of_3)','TAKEN_TO_HOSPITAL','RELATED_FACTOR_(1)-PERSON_LEVEL','INJURY_SEVERITY'])
```
